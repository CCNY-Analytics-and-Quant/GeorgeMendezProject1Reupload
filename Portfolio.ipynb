```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#part (a)

#download the 
tickers=['SKT','DHC','AVB','TSLA','MSFT','GOOGL','IBDRY'] # first 3 are REIT's, next 3 are tech companies, last 1 is green energy
etfs=['SPY','IWM','DIA']
df_my_final_project={'ticker': tickers}
df_my_final_project=pd.DataFrame(df_my_final_project)
```

```python
#my_portfolio (setting the weights)
weight=[1/7*100]*7 # 14 ownership in each stock
df_my_final_project.insert(1,'weight', weight, True)
```

```python
#code from class
adj_close = yf.download(tickers+etfs,period='10y')['Adj Close']
close = yf.download(tickers+etfs,period='10y')['Close']
returns = adj_close.pct_change()

data=(returns[-63:].std())*(252**.5)
datas=[]
for i in data:
    datas.append(i)
df_my_final_project['Volatility']=pd.DataFrame(datas)
```

```python
#Beta against spy
for i in etfs:
    data=returns[-252:].cov()['SPY']/returns[-252:]['SPY'].var()
    datas=[]
    for j in data:
        datas.append(j)
    df_my_final_project['Beta against '+i]=pd.DataFrame(datas)
df_my_final_project
```

```python
#Calculate 52 week high (52-week Low minus 52-week High) / 52-week High (According to the pdf)
listing=[]
for i in tickers:
    listing.append((min(adj_close[i][-366:-1])-max(adj_close[i][-366:-1]))/max(adj_close[i][-366:-1]))

df_my_final_project['Average Weekly Drawdown']=pd.DataFrame(listing)
min(2,6)
```

```python
#The formula for calculating total return is Total Return = (Ending Value â€“ Beginning Value + Dividends or Interest) / Beginning Value * 100
#link: https://www.wallstreetmojo.com/total-return-formula/ (i googled "total return formula")
listing=[]
for i in tickers:
    listing.append((adj_close[i][-1]-adj_close[i][0])/adj_close[i][0]*100)
df_my_final_project['Total Return']=pd.DataFrame(listing)
```

```python
#link: https://www.investopedia.com/terms/a/annualized-total-return.asp#:~:text=Key%20Takeaways-,An%20annualized%20total%20return%20is%20the%20geometric%20average%20amount%20of,the%20annual%20return%20was%20compounded.
#AR=(1+cumutalive_returns)^(365/days_held)-1    
#link: https://www.wallstreetoasis.com/resources/skills/finance/annualized-total-return final/initial^1/n-1
listing=[]
for i in tickers:
    listing.append(((adj_close[i][-1]/adj_close[i][0])**.1)-1)
df_my_final_project['Annualized Total Return']=pd.DataFrame(listing)
```
